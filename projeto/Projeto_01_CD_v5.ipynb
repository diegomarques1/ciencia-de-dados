Iremos também analisar se existem outliers na base e se vale a pena removê-los. Primeiro, será analisado o atributo mandante_Placar.

````python
tamanho = len(df)
primeiro_quartil = df.quantile(q=0.25, axis=0, interpolation='linear', numeric_only=True)
terceiro_quartil = df.quantile(q=0.75, axis=0, interpolation='linear', numeric_only=True)
````

````python
tamanho
````

````python
primeiro_quartil
````

````python
terceiro_quartil
````

````python
FIQ = terceiro_quartil.mandante_Placar - primeiro_quartil.mandante_Placar
FIQ
````

````python
# Usa-se a regra 1,5xFIQ
menor = primeiro_quartil.mandante_Placar - (int(FIQ) * 1.5)
maior = (int(FIQ) * 1.5) + terceiro_quartil.mandante_Placar

print("Maior: ", maior)
print("\nMenor: ", menor)
````

````python
# De acordo com o resultado, qualquer valor abaixo será um outlier
len(df[(df.mandante_Placar > menor)])
````

````python
# De acordo com o resultado, qualquer valor acima será um outlier
len(df[(df.mandante_Placar < maior)])
````

Podemos ver que, segundo o cálculo do intervalo interquartil, existem outliers para goleadas com mais de 3.5 gols de mandantes. Não existem jogos com gols negativos, então não existem outliers com -0.5 gols (não faz sentido mesmo).

````python
new_data = []
indice_lista = []

for ano, group in grouped:
    group_by_result = {}

    for indice, linha in group.iterrows():
        indice_lista.append(indice)

        # Inicializar dicionário para o mandante, se ainda não existir
        if linha["mandante"] not in group_by_result:
          group_by_result[linha["mandante"]] = {
              "gols_feitos_em_casa": 0,
              "gols_feitos_fora": 0,
              "gols_sofridos_em_casa": 0,
              "gols_sofridos_fora": 0,
              "vitorias_em_casa": 0,
              "empates_em_casa": 0,
              "derrotas_em_casa": 0,
              "vitorias_fora": 0,
              "empates_fora": 0,
              "derrotas_fora": 0,
          }

        # Inicializar dicionário para o visitante, se ainda não existir
        if linha["visitante"] not in group_by_result:
          group_by_result[linha["visitante"]] = {
              "gols_feitos_em_casa": 0,
              "gols_feitos_fora": 0,
              "gols_sofridos_em_casa": 0,
              "gols_sofridos_fora": 0,
              "vitorias_em_casa": 0,
              "empates_em_casa": 0,
              "derrotas_em_casa": 0,
              "vitorias_fora": 0,
              "empates_fora": 0,
              "derrotas_fora": 0,
          }

        new_data.append({
          "mandante_gols_feitos_em_casa": group_by_result[linha["mandante"]]["gols_feitos_em_casa"],
          "mandante_gols_feitos_fora": group_by_result[linha["mandante"]]["gols_feitos_fora"],
          "mandante_gols_sofridos_em_casa": group_by_result[linha["mandante"]]["gols_sofridos_em_casa"],
          "mandante_gols_sofridos_fora": group_by_result[linha["mandante"]]["gols_sofridos_fora"],
          "mandante_vitorias_em_casa": group_by_result[linha["mandante"]]["vitorias_em_casa"],
          "mandante_empates_em_casa": group_by_result[linha["mandante"]]["empates_em_casa"],
          "mandante_derrotas_em_casa": group_by_result[linha["mandante"]]["derrotas_em_casa"],
          "mandante_vitorias_fora": group_by_result[linha["mandante"]]["vitorias_fora"],
          "mandante_empates_fora": group_by_result[linha["mandante"]]["empates_fora"],
          "mandante_derrotas_fora": group_by_result[linha["mandante"]]["derrotas_fora"],
          "visitante_gols_feitos_em_casa": group_by_result[linha["visitante"]]["gols_feitos_em_casa"],
          "visitante_gols_feitos_fora": group_by_result[linha["visitante"]]["gols_feitos_fora"],
          "visitante_gols_sofridos_em_casa": group_by_result[linha["visitante"]]["gols_sofridos_em_casa"],
          "visitante_gols_sofridos_fora": group_by_result[linha["visitante"]]["gols_sofridos_fora"],
          "visitante_vitorias_em_casa": group_by_result[linha["visitante"]]["vitorias_em_casa"],
          "visitante_empates_em_casa": group_by_result[linha["visitante"]]["empates_em_casa"],
          "visitante_derrotas_em_casa": group_by_result[linha["visitante"]]["derrotas_em_casa"],
          "visitante_vitorias_fora": group_by_result[linha["visitante"]]["vitorias_fora"],
          "visitante_empates_fora": group_by_result[linha["visitante"]]["empates_fora"],
          "visitante_derrotas_fora": group_by_result[linha["visitante"]]["derrotas_fora"],
        })

        if linha['resultado'] == 0:
          group_by_result[linha["mandante"]]["vitorias_em_casa"] += 1
          group_by_result[linha["visitante"]]["derrotas_fora"] += 1
        elif linha['resultado'] == 1:
          group_by_result[linha["mandante"]]["derrotas_em_casa"] += 1
          group_by_result[linha["visitante"]]["vitorias_fora"] += 1
        else:
          group_by_result[linha["mandante"]]["empates_em_casa"] += 1
          group_by_result[linha["visitante"]]["empates_fora"] += 1

        group_by_result[linha["mandante"]]["gols_feitos_em_casa"] += linha["mandante_placar"]
        group_by_result[linha["visitante"]]["gols_feitos_fora"] += linha["visitante_placar"]
        group_by_result[linha["mandante"]]["gols_sofridos_em_casa"] += linha["visitante_placar"]
        group_by_result[linha["visitante"]]["gols_sofridos_fora"] += linha["mandante_placar"]
````
